<h2> 📌 개요 </h2>

<h4>1-1. 문제 설명</h4>

2024년에 주목받는 사회 이슈는 <Strong>"저출산"</Strong>으로, 최근 발표된 통계 자료를 통해 이 문제의 심각성이 더욱 부각되고 있다.

최근 발표된 통계자료에 따르면 합계출산율이 지속적으로 하락하고 있음을 보여준다. 2023년의 합계출산율은 0.72명으로, 2015년 이후 계속해서 감소했다.

이 문제는 인구 감소뿐만 아니라 경제적 부담도 초래할 수 있음을 인식하게 되었다. 출산율의 저하로 인한 인구 고령화 문제는 사회 전반에 영향을 미치며, 경제적인 측면에서도 심각한 영향을 미칠 수 있다. 예를 들어, 노동 인구의 감소는 생산성과 소비를 저하시키고 사회보장체계에 부담을 줄 수 있다.

이에 더해, 최근의 소득과 소비의 변화가 사회 구조와 경제에 큰 영향을 미치고 있음이 인식되고 있다. 소득격차의 증가와 소비 패턴의 변화는 가계의 경제적 안정성을 영향을 줄 수 있으며,이는 결국 가족의 출산 의욕에도 영향을 미칠 수 있다. 경제적으로 안정된 가정에서는 출산을 포함한 가족 활동에 대한 결정이 더 쉽게 이루어질 수 있기 때문이다.

따라서 최근의 경제적 변화를 조사하여 연도별 소득과 소비의 평균을 비교하고, 이것이 출산율에 부정적인 영향을 미치는지에 대한 연구가 필요하다는 생각이 들었으며, 이로인해 <strong> "소득과 소비에 따른 경제적 변화가 출산율에 미치는 영향"</strong>을 프로젝트 주제로 선정하게 되었다.

<h4>1-2. 분석의 목적 </h4>

    1) 한국의 출산율 감소 현상을 경제적 요인과의 연관성을 조사한다. 최근 한국은 출산율이 계속해서 하락하고 있어 인구 고령화와 관련된 여러 사회경제적 문제를 유발하고 있다. 이에 따라 평균 소득과 지출이 출산율에 미치는 영향을 분석하여, 경제적 여유가 출산 의사결정에 미치는 영향을 파악한다.
    2) 분석은 "통계청(KOSIS)"과 "나라지표"에서 제공하는 데이터를 기반으로 수행한다. 가계수지 데이터를 통해 소득과 지출의 변화를 살펴보고, 출산율 데이터를 통해 출산율의 변동 패턴을 분석하며,. 이러한 데이터를 통해 출산율과 경제적 요인 간의 관계를 파악하고, 어떤 요인이 출산율에 영향을 파악한다.
    3) 데이터 전처리 및 탐색 단계에서는 결측치 처리, 이상치 처리, 데이터 정규화/표준화 등을 수행한다. 또한, 기술통계 분석과 데이터 분포 확인을 통해 데이터의 특성을 파악하고, 상관관계 분석을 통해 변수 간의 관련성을 탐색하며, 중요 변수를 시각화하여 데이터의 인사이트를 도출한다.

<h4>2-3. 수집된 데이터의 구조 및 특성 설명</h4>

    소득 구간별로 세분화된 정보를 제공하여 소득 수준에 따른 가계의 경제 상태를 파악할 수 있다.

    출산율의 변화 추이와 연령별 출산율의 패턴 등을 분석할 수 있다.

    이러한 데이터를 통해 한국의 출산율과 가계의 경제 관계를 분석하여, 출산율 감소 현상에 대한 근본적인 원인을 파악하고자 한다.

    식료품 · 비주류음료: 연도별로 증감이 있지만 전반적으로 증가하는 추세를 보입니다. 특히 2019년 대비 2020년에 큰 증가가 있었으며, 이후에는 비슷한 수준을 유지하고 있습니다.

<strong>3. 가구원수와 가계지출의 관계 분석</strong>:

    가구원수가 증가함에 따라 가계지출이 어떻게 변화하는지를 살펴보면, 일반적으로 가구원수가 많을수록 가계지출이 증가하는 경향을 보입니다. 이는 가구원 수가 증가하면 가구 내에서 필요한 지출 항목이 더 많아지기 때문입니다.
    
<strong>4. 소득 구간별 가계지출 비율 분석</strong>:

    각 소득 구간별로 가계지출의 항목별 비율을 살펴보면, 소득 수준이 높아질수록 특정 항목에 대한 비중이 달라지는 경향을 보입니다. 예를 들어, 고소득 가구에서는 교육에 대한 지출 비중이 높은 반면, 저소득 가구에서는 식료품에 대한 비중이 더 높은 것을 확인할 수 있습니다.
    
<strong>5. 가계수지 항목별 소비 지출의 총액 변화 분석</strong>:

    각 가계수지 항목별로 연도별 소비 지출의 총액 변화를 살펴보면, 특정 항목에서 소비가 늘어나거나 줄어든 원인을 파악할 수 있습니다. 예를 들어, 의류 · 신발 항목에서는 2019년 대비 2020년에 감소한 후 2021년에 증가한 것을 확인할 수 있습니다.

<h4>4-4. 중요 변수 시각화</h4>

소비와 소득, 그리고 합계출산율 데이터를 시각화하여 분석할 예정이다.
막대그래프를 활용하여 각 연도별로 소비와 소득의 변화를 비교하고, 선 그래프를 사용하여 연도별 합계출산율의 추이를 살펴볼 예정이다.

막대그래프는 각 연도별 소비와 소득의 수준을 한눈에 파악할 수 있도록 도와주며, 선 그래프는 연도별 합계출산율의 추이를 추적하여 어떠한 경향이 있는지를 분석할 수 있도록 한다. 이를 통해 소비와 소득의 변화가 출산율에 미치는 영향을 시각적으로 파악할 예정이다.
